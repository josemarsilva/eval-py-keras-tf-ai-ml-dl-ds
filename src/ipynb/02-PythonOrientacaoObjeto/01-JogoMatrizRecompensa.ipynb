## 2. Classe, construtor, self e herança

```python
class Robo:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def print_x(self):
        print('x:', self.x) # imprime coordenada x

robo1 = Robo(5,5)
robo1.print_x()
print('type of robo1 is ', type(robo1))

class Robo3D(Robo): # Digo entre parenteses a superclasse da qual herdo
    def __init__(self, x, y, z): # tenho um novo construtor
        super(Robo3D,self).__init__(x,y) # ainda posso invocar o construtor da super classe
        self.z = z

robo2 = Robo3D(5,6,7)
robo2.print_x()
print('type of robo2 is ', type(robo2))

#robo2.print_x # Robo3D herda definições da classe Robo


```

## 3. Dando movimento ao robo

```python
class Ponto:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return '<%s, %s>' % (self.x, self.y)   
```

```python
class Recompensa(Ponto):
    
    def __init__(self, x, y, nome):
        super(Recompensa, self).__init__(x, y)
        self.nome = nome
        
    def __str__(self):
        return '<%s, %s, %s>' % (self.x, self.y, self.nome)
    
    def __repr__(self):
        return self.__str__()
```

```python
recompensa1 = Recompensa(1,1,'moeda de ouro')
print(recompensa1)
recompensa1
```

```python
class Robo:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return '<%s,%s>' % (self.x, self.y)

    def move_up(self):
        if self.y < 10:
            self.y = self.y + 1
        else:
            print('movimento(y+1) invalido!')

    def move_down(self):
        if self.y > 0:
            self.y = self.y - 1
        else:
            print('movimento(y-1) invalido!')

    def move_right(self):
        if self.x < 10:
            self.x = self.x + 1
        else:
            print('movimento(x+1) invalido!')

    def move_left(self):
        if self.x > 0:
            self.x = self.x - 1
        else:
            print('movimento(x-1) invalido!')

r1 = Robo(5,5)
print('Robo r1 (%i,%i)' % (r1.x, r1.y))
print(r1)
```

```python
def check_recompensa(robo, recompensas):
    retorno = False
    for recompensa in recompensas:
        if recompensa.x == robo.x and recompensa.y == robo.y:
            print('robo achou a recompensa %s' % recompensa.nome) # mostra todas as recompensas que o robo pode ter pego
            retorno = True
    return retorno
```

```python
recompensa1 = Recompensa(1, 2, 'moeda')
recompensa2 = Recompensa(5, 3, 'gasolina')
recompensas = [recompensa1, recompensa2]
print(recompensa1)
print(recompensa2)
robo = Robo(2,2)
print (check_recompensa(robo,recompensas))
```
